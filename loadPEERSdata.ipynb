```python
### memory task full scenario :

subject = 63
session_max = 6
list_max = 15

for i in presents:
    subject_index = i[0][0]
    session_index = i[0][1]
    list_index = i[0][3]
    if (subject_index == subject) and (session_index in range(1,session_max+1)) and (list_index in range(1,list_max+1)):
        print("STUDY PHASE =",subject_index,session_index,list_index)
        for x in i:
            print(x[5])
        print(".."*20)
        for r in recalls:
            subject_index_r = r[0][0]
            session_index_r = r[0][1]
            list_index_r = r[0][3] 
            if (subject_index_r == subject) and (session_index_r == session_index) and (list_index_r == list_index):
                print("RECALL PHASE =",subject_index_r,session_index_r,list_index_r)
                for y in r:
                    print(y[5],"--------->",y[4])
                break
    print("=="*20,"\n")

# for s in range(1,session_max+1):
#     for l in range(1,list_max+1):
#     print(presents[i])
#     print("\n\n")
```
